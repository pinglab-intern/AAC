#### Reactome large data sets

```python
rdf = pd.read_csv("reactome-data/pathways.csv")
```

```python
rdf.columns
```

```python
DATA = []
for ID,name,entities in zip(rdf['Pathway identifier'],rdf['Pathway name'],rdf['Submitted entities found']):
    #print(ID,name,entities)
    entities = entities.split(";")
    for p in entities:
        DATA.append({"ID": ID, "name": name, "protein": p})
```

```python
DATA[0]
```

```python
with open('reactome-data/pathway-data-large.json', 'w') as f:
    json.dump(DATA, f)
```
